## Tranforming the Extracted Data form .json format to Pandas DataFrames

### Aggregated Transations

```python
directory1 = "C:/Users/sadhi/Desktop/Phonepe Project/data_extracted/data/aggregated/transaction/country/india/state/"
A_transactionsList = os.listdir(directory1)

# Intialising a Dictionary that will be used to form a Pandas DataFrame and contains the column names of that DataFrame
col1 = {
    'State': [], 
    'Year': [], 
    'Quarter': [], 
    'Transaction_type': [], 
    'Transaction_count': [],
    'Transaction_amount': []
}


# Using nested for Loops to access the files in a particular directory (Phonepe data)
# state --> year --> file
for state in A_transactionsList:
    current_state = directory1 + state + "/"
    A_yearList = os.listdir(current_state)
    
    for year in A_yearList:
        current_year = current_state + year + "/"
        A_fileList = os.listdir(current_year)
        
        for file in A_fileList:
            current_file = current_year + file
            data1 = open(current_file, 'r')
            F1 = json.load(data1)
            
            # Extracting the relevent data from the .json file and appending it into the list inside the dictionary col1
            for i in F1['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                col1['State'].append(state)
                col1['Year'].append(year)
                col1['Quarter'].append(int(file.strip('.json')))
                col1['Transaction_type'].append(name)
                col1['Transaction_count'].append(count)
                col1['Transaction_amount'].append(amount)
                
df_A_Transactions = pd.DataFrame(col1)
print(df_A_Transactions.info())
df_A_Transactions.head()
```

### Aggregated Users

```python
directory2 = "C:/Users/sadhi/Desktop/Phonepe Project/data_extracted/data/aggregated/user/country/india/state/"
A_usersList = os.listdir(directory2)

# Intialising a Dictionary that will be used to form a Pandas DataFrame and contains the column names of that DataFrame
col2 = {
    'State': [], 
    'Year': [], 
    'Quarter': [], 
    'Brands': [], 
    'Count': [],
    'Percentage': []
}


# Using nested for Loops to access the files in a particular directory (Phonepe data)
# state --> year --> file
for state in A_usersList:
    current_state = directory2 + state + "/"
    A_yearList = os.listdir(current_state)
    
    for year in A_yearList:
        current_year = current_state + year + "/"
        A_fileList = os.listdir(current_year)
        
        for file in A_fileList:
            current_file = current_year + file
            data2 = open(current_file, 'r')
            F2 = json.load(data2)
            
            # Extracting the relevent data from the .json file and appending it into the list inside the dictionary col2
            try:
                for i in F2['data']['usersByDevice']:
                    brand_name = i["brand"]
                    counts = i["count"]
                    percents = i["percentage"]
                    col2["State"].append(state)
                    col2["Year"].append(year)
                    col2["Quarter"].append(int(file.strip('.json')))
                    col2["Brands"].append(brand_name)
                    col2["Count"].append(counts)
                    col2["Percentage"].append(percents)
            except:
                pass
            
df_A_User = pd.DataFrame(col2)
print(df_A_User.info())
df_A_User.head()
```
